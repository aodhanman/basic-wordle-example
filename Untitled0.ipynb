```python
import random

# List of words to choose from
words = ["apple", "banana", "cherry", "orange", "grape", "pear"]

# Choose a random word from the list
chosen_word = random.choice(words)

# Set the number of guesses allowed
max_guesses = 6

# Set the list of allowed letters
allowed_letters = "abcdefghijklmnopqrstuvwxyz"

# Initialize the guessed letters list
guessed_letters = []

# Game loop
while max_guesses > 0:
    # Print the current state of the game
    print(" ".join([letter if letter in guessed_letters else "_" for letter in chosen_word]))
    print(f"Guesses left: {max_guesses}")
    print(f"Guessed letters: {' '.join(guessed_letters)}")
    
    # Get the user's guess
    guess = input("Guess a letter: ").lower()
    
    # Check if the guess is valid
    if guess not in allowed_letters:
        print("Invalid guess. Please enter a valid letter.")
        continue
    elif guess in guessed_letters:
        print("You already guessed that letter. Please try again.")
        continue
    
    # Add the guess to the guessed letters list
    guessed_letters.append(guess)
    
    # Check if the guess is in the chosen word
    if guess in chosen_word:
        print("Correct!")
        
        # Check if the user has guessed all the letters
        if set(chosen_word) == set(guessed_letters):
            print("Congratulations! You guessed the word!")
            break
    else:
        print("Incorrect.")
        max_guesses -= 1
        
# Check if the user ran out of guesses
if max_guesses == 0:
    print(f"Sorry, you ran out of guesses. The word was {chosen_word}.")
```
